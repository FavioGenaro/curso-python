# Gráfico de Barras con Líneas
Vamos a combinar un gráfico de barras con un gráfico de líneas.
Las línas estan delimitadas por un segundo eje Y.



### Cuando utilizar
- Quiero comparar cambios de magnitudes de una variable a través del tiempo y añadir una segunda variable como un gráfico de líneas que tenga una escala diferente


### Consejos
- Los mismos consejos que el gráfico de barras y de līneas.
- Como consejo general, no recomiendo añadir más de dos líneas al gráfico. Se volvería muy complejo de leer.


## En esta clase
### Datos
Información sobre paises por año: Expectativa de vida, Población y GDP per Capita
### Objetivos
1. Comparar la expectativa de vida promedio en Africa a lo largo de los años y al mismo tiempo ver como ha cambiado el GDP per Capita.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
#sns.set_palette("colorblind")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/paises_vida.csv', index_col='index')
df
```

```python
# Eliminamos filas NaNs, agrupamos por año y por continente
df = df.dropna()
# solo de África
df_agrupado = df[df['continent'] == 'Africa'].groupby(['year']).agg({ # agrupamos por año
    'lifeExp': np.mean, # expectativa de vida
    'pop': np.sum, # población
    'gdpPercap': np.sum # gdp
})
df_agrupado = df_agrupado.reset_index() # quitar el indice
df_agrupado
```

```python
# Empecemos con el grafico de barras
sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey' # todas las barras de un mismo color
    # hue='continent'
)
```

```python
# Datos para el grafico de linea
# df_agrupado_pop = df[[]].groupby(['year']).agg({
#     'pop': np.sum,
# })
# df_agrupado_pop = df_agrupado_pop.reset_index()
# df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3)) # instanciamos un canvas

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    color='grey',
    alpha=0.8,
    ax=ax1 # se dibuja el barplot sobre el ax1
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx() # con ax2 graficamos sobre el ax1 que tiene el gráfico de barras

line_color = 'brown'
ax2.set_ylabel('Población', color=line_color)
sns.lineplot(
    data=df_agrupado['pop'], # seria el segundo eje Y. el eje X es compartido
    alpha=1,
    marker='D', # marcador en rombo
    ax=ax2,
    linewidth=2,
    color=line_color
)

# color de los valores del segundo eje Y
ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```

```python
df_agrupado = df.groupby(['year', 'continent']).agg({
    'lifeExp': np.mean,
    'pop': np.sum,
    'gdpPercap': np.sum
})
df_agrupado = df_agrupado.reset_index()
df_agrupado
```

```python
# Datos para el grafico de linea
df_agrupado_pop = df.groupby(['year']).agg({
    'pop': np.sum,
})
df_agrupado_pop = df_agrupado_pop.reset_index()
df_agrupado_pop
```

```python
fig, ax1 = plt.subplots(figsize=(7,3))

sns.barplot(
    data=df_agrupado,
    x='year',
    y='lifeExp',
    alpha=0.8,
    hue='continent', # para generar columnas multiples
)
ax1.set_axisbelow(True)

ax2 = ax1.twinx()

line_color = 'brown'
ax2.set_ylabel('Población', color=line_color)
sns.lineplot(
    data=df_agrupado_pop['pop'],
    alpha=1,
    marker='D',
    ax=ax2,
    linewidth=2,
    color=line_color
)

# color de los valores del segundo eje Y
ax2.tick_params(axis='y', labelcolor=line_color)

ax2.set_axisbelow(True)
ax2.grid(visible=False)

plt.show()
```
