# Joint Plot
Es un grfico compuesto por dos tipos de grficos diferentes: Es un Scatter Plot que al mismo tiempo muestra histogramas (o KDE) con la distribucin de cada variable. Para este grfico, ambas variables deben ser numricas.    
**Eje X**: Variable numrica  
**Eje Y**: Variable numrica  



## En esta clase
### Datos
Informacin sobre especies de pinguinos
### Objetivos
1. Encontrar alguna correlacin entre las caractersticas de los pinguinos y al mismo tiempo ver la distribucin de las variables

## Referencias
- https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    # cada histograma se grafica sobre el eje de cada varaible. X esta sobre el grafico y Y a la derecha de grafico
    x='flipper_length_mm', # aletas, hay bimodalidad. Vemos que por cada especie hay una campana
    y='culmen_length_mm', # pico, tambien tiene bimodalidad
    hue='species',
    # kind="reg" # muestra la linea de regresin y linea de tendencia sobre los histogramas
)
# muestran los histogramas
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species'
)
```
