# Heatmap (Mapa de Calor)
Es un grfico X vs Y con dos usos diferentes dependiendo del tipo de las variables en los ejes.   


**Si las dos variables son numricas**: El heatmap es una matriz que representa como se distribuyen las dos variables viendolas como un par de puntos. Es decir, que par de valores (x, y) tiene una mayor o menor frecuencia. 

Lo podemos ver como un Scatter Plot agrupado en cuadrculas equidistantes. En donde una cuadrcula con una tonalidad mas oscura me indica que existen mas de esos pares de puntos (x, y) dentro de esa cuadrcula. Y una cuadrcula con una tonalidad mas clara me indica que existen menos pares de puntos (x, y) dentro de esa cuadrcula. 

Por lo general, se lo utiliza cuando tengo un Scatter Plot con demasiados puntos o cuando quiero tener informacin sobre la distribucin de las variables como un par. Y no individualmente.

**Otros nombres**: Si las cuadrculas son hexagonales, se le denomina Hexabin. Y la funcionalidad es la misma.




**Si las dos variables son categricas y codifican una variable numrica**: El Heatmap es una matrz con tantas celdas como valores distintos categoricos tenga en X y en Y. Una celda con una tonalidad mas oscura me indica que el valor de ese punto X vs Y es mayor. **Ejemplos de Uso**: Como dueo de un supermercado me gustaria saber que tipo de productos se han vendido ms y en cuales supermercados. En este caso mi eje X serian los tipos de productos, mi eje Y serian los nombres de los supermercados, y la celda con su intensidad de color codificarian el valor de ventas de ese producto en ese supermercado.




**Otros Nombres**: Existe algo denominado Heatmap de Correlaciones o Matriz de Correlacin. En donde ambos ejes tienen nombres de variables (o columnas). Y la celda muestra el coeficiente de correlacin entre esas dos variables.




Otro ejemplo de un Heatmap seria el grafico del nmero de contribuciones de un usuario de Github. Rpidamente podemos ver que dias hemos hecho mas contribuciones. Es un grfico muy intuitivo y til para mostrar una cantidad grande de informacin de forma efectiva.




### Cuando utilizar
- Quiero observar la distribucin de dos variables como un par de puntos.
- Quiero visualizar de forma intuitiva mucha informacin de una gran cantidad de categoras en un solo grfico.

### Consejos
- Usualmente la tonalidad mas oscura representar un mayor valor.
- Es recomendable aadir una leyenda que muestre el significado de las tonalidades. Los rangos de valores que cada tonalidad codifica.
- Cuando utilizamos variables categricas es comn mostrar el valor que codifica la celda dentro de la celda.
- Es una alternativa cuando quiero hacer un grfico de barras mltiples, y tengo muchas categorias. Como el ejemplo que mencionamos. Si tengo muchos supermercados, o muchos tipos de productos diferentes. Un grfico de barras sera demasiado denso.
- Si tengo muy pocos puntos en mi dataset, es preferible un scatter plot. O si tengo solo una o dos categorias es mejor un diagrama de barras.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
# Flipper lenght vs Culmen Lenght 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm'
)
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')
```

```python
# Heatmap de datos numricos
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df, 
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='hist', # hist, tipo histograma que genera el heatmap. hex para un hexabin, debemos quitar el parmetro cbar
    joint_kws={
        'bins': 25, # aumentamos la cantidad de bins para hacer los cuadros ms pequeos
        #'gridsize': 30 # Para hexabin
    },
    marginal_kws={ # permite configurar los grficos a los lados (las barras)
        "bins":20, # 20 barras
        "fill":True
    },
    cbar=True # aade una leyenda en modo barra de colores
    #gridsize=30  # Para el tamao de cada hexabin
)
plt.show()
```

```python
# Hexabin de datos numricos
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.hexbin.html
# con mplt
df.plot.hexbin(
    x='flipper_length_mm',
    y='culmen_length_mm',
    gridsize=30
)
plt.grid(False)
plt.show()
```

# Heatmap de correlacin

```python
agrupado = df.groupby(['species', 'island']).count()[['sex']]
agrupado
```

```python
# revertimos los ejes
pivote = agrupado.unstack().droplevel(0, axis=1)
pivote
```

```python
# Heatmap de variables categricas que codifican una variable numrica
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
sns.heatmap(
    pivote,
    annot=True,
    cmap="crest",
    linewidth=0.5,
    fmt=".1f"
)
# se tiene el coeficiente de correlacin en cada recuadro
```

```python
sns.pairplot(
    df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]
)
```

```python
# Heatmap de Correlacin
# https://seaborn.pydata.org/generated/seaborn.heatmap.html

# Calculo los coeficientes de correlacion lineal entre mis variables
df_cor = df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].corr()
sns.heatmap(
    df_cor,
    annot=True,
    cmap='coolwarm_r', # Blues
    square=True # cuadrados perfectos, no rectangulos
)

palette = sns.color_palette("Blues_d",n_colors=15)
palette.reverse()
```

# Heatmap categrico

```python
df2 = pd.read_csv('../data/tienda_ventas.csv')
df2 = df2.dropna()
# creamos la columna periodo
df2['periodo'] = df2['date'].apply(lambda x: x[:7])
df2
```

```python
# filtramos la lista a cierta familia y ciertas tiendas
tiendas_agrupadas = df2[(df2['family'] == 'BEVERAGES') & (df2['store_nbr'].isin([3, 44, 45, 46, 47, 48, 49, 50]))].groupby(['periodo', 'store_nbr']).agg({
    'sales': np.sum # sumamos las ventas
})
tiendas_agrupadas
```

```python
# intercambiamos filas y columnas (transpose, transpone la matriz) y eliminamos un nivel
pivote2 = tiendas_agrupadas.unstack().droplevel(0, axis=1).transpose()
pivote2
```

```python
# Segundo ejemplo de un Heatmap de dos variables categricas que codifican una variable numrica
plt.figure(figsize=(15, 8)) # tamao del grfico
ax = sns.heatmap(
    pivote2,
    square=True, # Me genera cuadrados simetricos
    cmap="Blues", # escala de colores
    linewidth=0.1, # ancho de la linea que separa los cuadrados
    # annot=True # agrgar los valores dentro de los cuadrados, pero al ser valores muy grandres chocan unos a otros
)
plt.show()
```

```python
tiendas_agrupadas_2 = df2.groupby(['family', 'store_nbr']).agg({
    'sales': np.sum
})
tiendas_agrupadas_2
pivote3 = tiendas_agrupadas_2.unstack().droplevel(0, axis=1).transpose()
pivote3 = (pivote3-pivote3.min())/(pivote3.max()-pivote3.min())
pivote3
```

```python
# Segundo ejemplo de un Heatmap de dos variables categricas que codifican una variable numrica
plt.figure(figsize=(15, 5))
ax = sns.heatmap(
    pivote3,
    square=False, # Me genera cuadraditos simetricos
    cmap="Blues",
    linewidth=0.1,
)
plt.show()
```

```python
# Intentando hacer un grafico de barras multiples con la misma informacion
plt.figure(figsize=(15, 5))
temp_df = tiendas_agrupadas_2.reset_index()
temp_df
ax = sns.barplot(
    temp_df,
    x='family',
    y='sales',
    hue='store_nbr',
    # color='red',
    linewidth=0
)
ax.set_xticklabels(
    set(temp_df['family']),
    rotation=90 # Rotacion de las etiquetas
)
plt.show()
```
