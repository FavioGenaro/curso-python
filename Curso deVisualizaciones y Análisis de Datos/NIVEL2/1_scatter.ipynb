# Grfico de Dispersin (Scatter Plot)
Grfico que muestra los valores de dos variables de un mismo registro en nuestro dataset al mismo tiempo. En cada Eje del grfico se coloca una variable. Se dibuja en el canvas un punto, en la coordenada correspondiente al valor de las variables por cada observacin de nuestro dataset. Su objetivo es detectar si existe o no alguna correlacin entre las variables en los ejes. Estos ejes usualmente son variables numricas pero tambin se pueden utilizar **variables categricas ordinales**.   
**Eje X**: Variable numrica o categrica   
**Eje Y**: Variable numrica o categrica  
**Otros nombres**: Grfico de puntos, X-Y Plot. 



**Ejemplos de Uso**:
- Como un doctor, me gustara saber si el nivel de colesterol de una persona se correlaciona con su peso.

### Cuando utilizar
- Quiero comprobar si existe algun tipo de correlacin entre dos variables (ver imagen de tipos de correlaciones) 
- Quiero observar la distribucin de dos variables al mismo tiempo.
- Quiero hallar el tipo, y el nivel (o fuerza) de correlacin entre mis variables (ver imagen de tipos de correlaciones)




### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero". En realidad, ambos ejes deberan empezar donde empieza el rango de sus datos.
- No utilices colores en los puntos innecesariamente.
- Puedo introducir una 3ra o 4ta variable al grfico con las siguientes tcticas:
  - Diferentes colores de cada punto para introducir una variable categrica.
  - Diferentes formas en los puntos para introducir una variable categrica.
  - Convertirlo en un grfico de burbujas para introducir una variable numrica (el tamao de las burbujas representa esa otra variable).
  - Jugar con la intensidad del color de cada punto para introducir una variable numrica.
- Si tengo muy pocos puntos en mi dataset, el resultado del grfico no ser muy til.
- Si hay una cantidad muy grande de puntos, el grfico se vuelve ilegible. Con apariencia de manchas por estar los puntos muy juntos.
  - Podemos coger una muestra, o jugar con la opacidad de los puntos para que todos se puedan notar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set_style("whitegrid")

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Flipper lenght vs Culmen Lenght 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm'
)
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')

# mientras aumente al tamao de las aletas, tambin lo hace la del pico
```

```python
# Flipper lenght vs Culmen Lenght vs Culmen Depth
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm',
    c='culmen_depth_mm' # Tercera variable como color
)
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')
# vemos que tambien aumenta el peso
```

```python
# Scatter Plot con Plotly
# https://plotly.com/python/line-and-scatter/
fig = px.scatter(
    df, 
    x='flipper_length_mm', 
    y='culmen_depth_mm', 
    title='Longitud de las Aletas vs Profundidad del Pico en Pinguinos',
    # color='sex',
    symbol='species', # 
    color='species' # tercera variable para los colores
)
fig.show()
```

# Grfico de Dispersin Categrico 
En Seaborn hay un tipo de grfico especial que me ayuda cuando quiero hacer un Scatter plot utilizando variables categricas como los ejes. Me es til si quiero determinar cuando una variable categrica puede estar relacionada a cambios en una variable numrica. Uno de los dos ejes tiene que ser una variable categrica y el otro una variable numrica.  
**Eje X**: Variable numrica o categrica   
**Eje Y**: Variable numrica o categrica  
**Otros nombres**: Category Plot 


## Referencias
- https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
# https://seaborn.pydata.org/generated/seaborn.catplot.html
# util para relacionar variables categoricas con numericas
sns.catplot(
    data=df, 
    x="flipper_length_mm", 
    y="species", # categorica
    hue="island" # agregar colores
)
# la especie Gento tiene tamao de aletas ms grandes
```
