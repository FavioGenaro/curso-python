# Gráfico de Dispersión (Scatter Plot)
Gráfico que muestra los valores de dos variables de un mismo registro en nuestro dataset al mismo tiempo. En cada Eje del gráfico se coloca una variable. Se dibuja en el canvas un punto, en la coordenada correspondiente al valor de las variables por cada observación de nuestro dataset. Su objetivo es detectar si existe o no alguna correlación entre las variables en los ejes. Estos ejes usualmente son variables numéricas pero también se pueden utilizar **variables categóricas ordinales**.   
**Eje X**: Variable numérica o categórica   
**Eje Y**: Variable numérica o categórica  
**Otros nombres**: Gráfico de puntos, X-Y Plot. 



**Ejemplos de Uso**:
- Como un doctor, me gustaría saber si el nivel de colesterol de una persona se correlaciona con su peso.

### Cuando utilizar
- Quiero comprobar si existe algun tipo de correlación entre dos variables (ver imagen de tipos de correlaciones) 
- Quiero observar la distribución de dos variables al mismo tiempo.
- Quiero hallar el tipo, y el nivel (o fuerza) de correlación entre mis variables (ver imagen de tipos de correlaciones)




### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero". En realidad, ambos ejes deberían empezar donde empieza el rango de sus datos.
- No utilices colores en los puntos innecesariamente.
- Puedo introducir una 3ra o 4ta variable al gráfico con las siguientes técticas:
  - Diferentes colores de cada punto para introducir una variable categórica.
  - Diferentes formas en los puntos para introducir una variable categórica.
  - Convertirlo en un gráfico de burbujas para introducir una variable numérica (el tamaño de las burbujas representa esa otra variable).
  - Jugar con la intensidad del color de cada punto para introducir una variable numérica.
- Si tengo muy pocos puntos en mi dataset, el resultado del gráfico no será muy útil.
- Si hay una cantidad muy grande de puntos, el gráfico se vuelve ilegible. Con apariencia de manchas por estar los puntos muy juntos.
  - Podemos coger una muestra, o jugar con la opacidad de los puntos para que todos se puedan notar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set_style("whitegrid")

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Flipper lenght vs Culmen Lenght 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm'
)
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')

# mientras aumente al tamaño de las aletas, también lo hace la del pico
```

```python
# Flipper lenght vs Culmen Lenght vs Culmen Depth
df.plot.scatter(
    x='flipper_length_mm',
    y='culmen_length_mm',
    c='culmen_depth_mm' # Tercera variable como color
)
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')
# vemos que tambien aumenta el peso
```

```python
# Scatter Plot con Plotly
# https://plotly.com/python/line-and-scatter/
fig = px.scatter(
    df, 
    x='flipper_length_mm', 
    y='culmen_depth_mm', 
    title='Longitud de las Aletas vs Profundidad del Pico en Pinguinos',
    # color='sex',
    symbol='species', # 
    color='species' # tercera variable para los colores
)
fig.show()
```

# Gráfico de Dispersión Categórico 
En Seaborn hay un tipo de gráfico especial que me ayuda cuando quiero hacer un Scatter plot utilizando variables categóricas como los ejes. Me es útil si quiero determinar cuando una variable categórica puede estar relacionada a cambios en una variable numérica. Uno de los dos ejes tiene que ser una variable categórica y el otro una variable numérica.  
**Eje X**: Variable numérica o categórica   
**Eje Y**: Variable numérica o categórica  
**Otros nombres**: Category Plot 


## Referencias
- https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
# https://seaborn.pydata.org/generated/seaborn.catplot.html
# util para relacionar variables categoricas con numericas
sns.catplot(
    data=df, 
    x="flipper_length_mm", 
    y="species", # categorica
    hue="island" # agregar colores
)
# la especie Gento tiene tamaño de aletas más grandes
```
