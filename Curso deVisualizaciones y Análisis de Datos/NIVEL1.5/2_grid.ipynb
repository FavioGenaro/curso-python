# Grid (Cuadrcula)
Lneas opacas que sirven como ayuda visual al lector para codificar los valores o puntos del grfico a los valores en los ejes. 
- Es un elemento opcional de una visualizacin. Pero usualmente es recomendable utilizarlo.
- Si utilizamos una plantilla, esta va a configurar automticamente los estilos de la cuadrcula. 
- Podemos configurar el estilo de las lneas del Grid, la opacidad, colores y en que eje aparecen.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Preprocesamiento de datos
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

# Pandas

```python
plt.style.use(
    'ggplot' 
)

# El metodo plot me retorna una referencia al objeto de mi grfico
# Este objeto del grafico, es un objeto propio de matplotlib
ax = df.plot.line(
    figsize=(15, 5) 
)
```

```python
ax = df.plot.line(
    figsize=(15, 5) 
)

# Luego yo puedo empezar a configurar propiedades del grfico, como el GRID
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html
ax.xaxis.grid( # esto para el eje x
    visible=True,
    linewidth=1.5, # Ancho de las lneas del Grid
    color='blue', # Colores predefinidos: https://matplotlib.org/stable/gallery/color/named_colors.html#css-colors
    alpha=0.2 # Opacidad: Porcentaje del 0 al 1
)
```

```python
ax = df.plot.line(
    figsize=(15, 5) 
)

ax.xaxis.grid( # esto para el eje x
    visible=True,
    linewidth=1.5, # Ancho de las lneas del Grid
    color='blue', # Colores predefinidos: https://matplotlib.org/stable/gallery/color/named_colors.html#css-colors
    alpha=0.2 # Opacidad: Porcentaje del 0 al 1
)

ax.yaxis.grid( # para el eje Y
    visible=True,
    linewidth=0.5,
    color='#873e23', # Puedo utilizar colores en HEX
    linestyle='dashed', # Estilos de Linea: https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle  
)

# podemos aplicar estilos de lnea a los que representan los valores.
```

# Seaborn

```python
# Dimensiones en pulgadas -> (ancho, alto)
plt.figure(
    figsize=(8,4)
)
sns.set_style("ticks")
sns.set_theme("paper")

# metodo global, aplica en todos los grficos que genere despues
# como Searborn usa plt, usar estas configuraciones. Plotly no usa plt
plt.grid( 
    visible=False,
    axis='x'
)

plt.grid(
    color='red',
    alpha=0.2,
    axis='y'
)

ax = sns.lineplot(
    data = df
)
```

```python
# Dimensiones en pulgadas -> (ancho, alto)
plt.figure(
    figsize=(10, 3)
)

# GRID: Metodo que funciona en Seaborn y Matplotlib
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

# Grid del Eje X
plt.grid(
    visible=True,
    axis='x',
    color='red',
    linestyle='--',
    alpha=0.5
)

# Grid del Eje Y
plt.grid(
    axis='y',
    color='blue',
    linestyle='--',
    alpha=0.5
)


sns.lineplot(
    df
)
```

```python
fig = px.line( # retornamos la imagen
    df, 
    width=800, 
    height=500,
    template='ggplot2'
)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(
    showgrid=True,
    gridwidth=1, # Ancho de las lneas en pixeles
    gridcolor='black', # Colores disponibles: https://developer.mozilla.org/en-US/docs/Web/CSS/named-color
    griddash='dot', # Estilos de lineas: "solid", "dot", "dash", "longdash", "dashdot", o "longdashdot"
    minor_griddash='dot', # Estilos de los ejes menores, crea ejes entre los ejes principales, pero ms opacos (el eje principal son los que estan en los valores)
    # No hay forma de cambiar la opacidad
)

fig
```
