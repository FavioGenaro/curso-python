# Anotaciones
A veces querremos escribir **notas** en el grfico o resaltar algn punto de inters con texto.



 Tambin es util utilizar **marcas** como lneas verticales u horizontales para resaltar lmites o umbrales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

# Pandas

```python
plt.style.use(
    'ggplot' 
)
ax = df.plot.line(
    figsize=(10, 4)
)

ax.set_title(
    'Ventas de 3 Tipos de Productos por Ao',
    fontdict={
        'size': 18,
        'name': 'Helvetica'
    }
)

ax.set_xlabel(
    'Perodo',
    fontdict={
        'name': 'Helvetica',
        'color': 'black',
        'size': 12
    }    
)


ax.set_ylabel(
    'Ventas (en Millones de Dlares)',
    fontdict={
        'name': 'Helvetica', # fuente
        'color': 'black',
        'size': 12
    }
)

ax.set_xticks(np.arange(0, len(df))) 
ax.set_xticklabels(
    df.index.values, 
    rotation=45,
    fontdict={
        'color':  'grey',
        'size': 10
    }
)

# Aadir anotaciones, sirve para matplotlib y seaborn
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
ax.annotate(
    'Noviembre', # El texto
    # los valores son los de eje X (posicin 10, considerando que se empieza de 0) y valor nmerico de Y (1 milln 250 mil)
    xy = (10, 1250000), # La ubicacin del punto sobre el cual hare mi anotacin (x, y). A este punto apuntara mi flecha
    xytext = (10, 1000000), # La ubicacion donde estara mi texto
    size=10, 
    color='grey',
    horizontalalignment='center',
    arrowprops={ # flecha, esto se debe agregar para que aparezca la flecha
        'width': 1,
        'color': 'black',
        'headwidth': 5, # ancho
        'headlength': 5 # largo de la cabeza
    }
)

plt.yticks(
    fontsize=10,
    color='grey',
    name='Helvetica'
)

plt.show()
```

# Seaborn

```python
plt.figure(
    figsize=(10, 4)
)

ax = sns.lineplot(
    df
)

ax.set(
    xlabel='Perodo', 
    ylabel='Ventas (en Millones de Dlares)',
    title='Ventas de 3 Tipos de Productos por Ao (Seaborn)'
)

# Marcar una lnea en el grfico
# Lineas horizontales (axhline): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.axhline.html
# Lineas verticales (axvline): https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.axvline.html
ax.axhline( # linea horizontal
    1400000, # Punto en el Eje Y donde quiero trazar mi linea horizontal
    linestyle='dashed',
    color='red',
    alpha=1,
    linewidth=0.7
)

# anotacin para explicar la lnea
ax.annotate(
    'Lmite', # El texto
    xy = (11, 1400000), # La ubicacin del punto sobre el cual hare mi anotacin (x, y). A este punto apuntara mi flecha
    xytext = (11, 1400000), # La ubicacion donde estara mi texto
    size=10, 
    color='red',
    # alineacin del texto
    horizontalalignment='center',
    verticalalignment='top',
    fontstyle='italic' # https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
)

plt.show()
```

# Plotly

```python
fig = px.line(
    df,
    width=800, 
    height=500,
    template='ggplot2',
    title='Ventas de 3 Tipos de Productos por Ao', 
    labels={
        "fecha": "Periodo",
        "value": "Ventas por ao (en millones de $)",
        "variable": "Tipo de Producto"
    }
)

fig.update_layout( 
    font_family="Rockwell",
    font_color="black"
)

fig.update_xaxes(
    tickangle=45,
    tickmode='array',
    tickvals=df.index.values,
    tickfont={
        'color': 'grey',
        'size': 10
    }
)

fig.update_yaxes(
    tickfont={
        'color': 'grey',
        'size': 10
    }
)

fig.update_yaxes(
    nticks=20
)

# Anotaciones en Plotly
# https://plotly.com/python/text-and-annotations/#styling-and-coloring-annotations
fig.add_annotation(
    x='2015-11', # permite usar el valor original
    y=1300000, # posicin segn el valor
    text="Noviembre",
    showarrow=True, # Linea de la flecha
    arrowhead=3, # Tipos de flecha: numero del 0 al 8, 0 es sin cabeza
    arrowsize=1, # Tamao de la cabeza de la flecha
    arrowwidth=2, # Ancho de la flecha en pixeles
    arrowcolor="grey",
    opacity=1,
    font={
        'family': "Arial",
        'size': 10,
        'color': "black"
    }
)

# Marcas en Plotly
# https://plotly.com/python/horizontal-vertical-shapes/#adding-text-annotations
fig.add_hline( # liena horizontal
    y=1400000, 
    line_dash="dot",
    annotation_text="Lmite", # Puedo incluir al mismo tiempo una anotacin
    annotation_position="top left", # "top left", "top right", "bottom left", "bottom", "top" (por defecto es center)
    annotation_font_size=10,
    annotation_font_color="red",
    opacity=0.8,
    line_width=1,
    line_color='red',
)

fig.show()
```
