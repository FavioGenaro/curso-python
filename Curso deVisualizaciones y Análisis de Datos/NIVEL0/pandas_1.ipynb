```python
import pandas as pd
```

```python
# CARGA DE DATOS
df = pd.read_csv("./dataset.csv") # Por defecto detecta las columnas por el encabezado del archivo
# Puedo especificar algunos parametros adicionales como por ejemplo una columna como indice
# pd.read_csv("./dataset.csv", index_col = "id")

df = df.set_index('id')
# DESCUBRIENDO LOS DATOS
df.tail(5)
df.describe()
df.columns
df.head(5)
```

```python
# LIMPIEZA DE DATOS

df_filtrado = df.dropna()
df_filtrado.head(5)

df_llenado = df.fillna(0)
df = df.fillna({"retweets": 0, "mentions": -1}) # reemplaza los valores NaN de cada columna por 0 y -1 respectivamente
df_llenado.head()
df.head()
```

```python
# SELECCION Y FILTRADO

# Seleccionando columnas, a traves de su nombre
df["favorites"] # Me devuelve una sola columna como una Serie
df["full_text"] # Me devuelve una sola columna como una Serie
df[["favorites", "full_text"]] # Me devuelve un dataframe solamente con dos columnas
```

```python
# Seleccionando filas, a traves del número de fila al que corresponde (indice de la columa)
df.iloc[0] # fila 1
df.iloc[0:2] # fila del primero a
df.iloc[[0, 1, 2]] # indicamos las filas 
```

```python
# Seleccionando filas, a traves de la columna tomada como index (en este caso con la columna id)
df.loc[183721] # fila de indice 0
df.loc[183721:183740] # fila del primero a
# df.loc[[183721, 183722, 183723]] # indicamos las filas 
```

```python
# Seleccionando filas y columnas al mismo tiempo
df.loc[[183721, 183722, 183723], ["favorites"]]
# Seleccionando todas las filas, y un grupo de columnas
df.loc[:, ["favorites", "full_text"]]
# Seleccionando un rango de filas y un grupo de columnas
df.loc[183721:183723, ["favorites", "full_text", "mentions"]]
```

```python
# Seleccionando por condicion
# Todas las filas que tengan un valor de favoritos mas altos de 100
df[df["favorites"] > 400] # de la columna favorito se toman los mayores a 400
df[(df["favorites"] > 400) & (df["mentions"] > 20)] # en pandas no una 'and' sino '&'
df[df["full_text"].str.contains("Programming")]
```

```python
# TRANSFORMACION DE DATOS
import random

# Agregando una columna de ganancias, calculada en base a los retweets
def calcularGanancias(retweets):
    return retweets * random.randint(3, 5)

df["ganancias"] = df["retweets"].apply(calcularGanancias) # nueva columna aplicando la función a la columna "retweets"
df.head(10)
```

```python
# Agregando una columna de popularidad, calculada en base a mi numero de seguidores y numero de pesonas que sigo
def popularidad(fila):
    return fila["followees"]/fila["followers"]

# para aplicarlo en varias columnas, 
df["popularidad"] = df.apply(popularidad, axis=1)
df.head(10)
```

```python
# AGRUPANDO Y AGREGANDO DATOS
# En recientes versiones de Pandas
# Tenemos que filtrar primero solo las columnas numericas y la columna por la que vamos a agrupar
df[ 
    ['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad'] # lista de columnas de df
].groupby("country").mean() # Agrupando por pais "country" y sacando la media de cada columna
```

```python
# Agrupando por pais y aplicando diferentes funciones de agregacion a cada columna
df.groupby("country").agg({ # funciones de agregación
    "followers": 'sum', # suma
    "mentions": 'mean', # media
    "retweets": 'max' # maximo valor
})

grouped = df.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})
grouped[grouped["followers"] > 5000] # filtramos las agrupación
```

```python
# ALMACENAMIENTO
grouped.to_csv("./salida.csv") # generamos un archivo csv
```
