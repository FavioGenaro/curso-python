# Estructuras de datos: Conjuntos



## Conjunto

Ahora, es el turno de los **conjuntos**. Éstos son una estructura sin orden que no admiten múltiples ocurrencias de un mismo elemento. Pueden ser definidos a partir de una lista con la función `set()`, o bien, los elementos del conjunto pueden ir entre llaves, `{}`, y estar separados por comas.

Los conjuntos son:
- hetereogéneos: los elementos pueden ser de distinto tipo en un mismo conjunto
- no mutables: los elementos no pueden ser modifcados una vez el conjunto ha sido creado

Los conjuntos pueden construirse con la función `set()` o directamente entre llaves, `{}`.

```python
# creamos un conjunto en base a una lista
set1 = set([1, 7, 4, 2, 0])
print(set1)
type(set1)
```

```python
# creamos una conjunto en base a una tupla
set2 = set((1, 7, 4, 2, 0))
print(set2)
type(set2)
```

```python
# declramos el conjunto directamente
# vemos que imprime los número en un orden comodo para el sistema, no tiene un orden especifico
set3 = {1, 7, 4, 2, 0}
print(set3)
type(set3)
```

Como se ha dicho antes, los conjuntos no admiten elementos repetidos:

```python
# no se pueden tener repetidos
set1 = {1, 2, 2, 3, 3, 3, 4, 4, 4, 4}
set1
```

**Observación.** Al decir que los conjuntos no tienen orden, lo que ocurre es que `Python` no mantendrá el que hemos introducido, tal y como hacía con las listas, sino que **reordenará todos los elementos por orden primero numérico (yendo antes los negativos que los positivos) y luego alfabético de las claves**. Con la función `print()`, si hay elementos negativos, el conjunto no se ordena correctamente.

```python
# el conjunto puede tener distintos tipos de datos
set1 = {1, "a", "i", "o", 2, "e", 4, 3, 5, "u"}
set1
```

```python
# El print cambia el orden
print(set1)
```

```python
set2 = {-1, "a", 3, "i", "o", 2, "e", 4, -3, 5, "u", 1}
set2
```

```python
print(set2)
```

```python
# un conjunto solo puede tener cadenas o numeros, pero no otros tipos de datos
set3 = {-1, "a", "i", "o", 2, "e", 4, -3, 5, "u", 1}
set3
```

Por lo que hemos visto hasta ahora, podemos decir que los conjuntos en `Python` son fieles a la definición matemática de éstos, salvo por el hecho de que en `Python` **un conjunto no puede contener como elemento a otro conjunto**:

**Conjunto.** Colección de elementos pertenecientes a la misma categoría, y cuya agrupación puede ser considerada o identificada en sí misma como un objeto. Un conjunto queda definido únicamente por sus elementos y por nada más. En particular, un conjunto puede escribirse como una lista de elementos, pero cambiar el orden de dicha lista o añadir elementos repetidos no define un conjunto nuevo.



### Subconjuntos

**Subconjunto.** Un subconjunto $B$ de un conjunto $A$ es un conjunto que contiene algunos de los elementos de $A$ (o quizá todos). Se denota por $B\subseteq A$

**Subconjunto propio.** Un subconjunto propio $B$ de un conjunto $A$ es un conjunto que tiene algunos de los elementos de $A$, pero no todos. Se denota por $B\subset A$

Para saber si un conjunto $B$ es subconjunto del conjunto $A$, podemos utilizar el método `.issubset()` o el comparador `<=`. Para saber si se trata de un subconjunto propio, tenemos el comparador `<`.

```python
A, B = {0, 3, 7, 2, 5}, {2, 3, 0}
# issubset determina si B es un subconjunto de A
B.issubset(A)
```

```python
# B es subconjunto de A, A y B pueden ser iguales
B <= A
```

```python
# B es subconjunto de A, en este caso A debe tener un elemento más que B
B < A
```

**Superconjunto.** Un superconjunto $A$ de un conjunto $B$ es un conjunto que contiene a $B$. Se denota por $A\supseteq B$. A incluye a B o son iguales. **B <= A**

**Superconjunto propio.** Un superconjunto propio $A$ de un conjunto $B$ es un conjunto que contiene a $B$ y consta de al menos un elemento más. Se denota por $A\supset B$. **B < A**

Para saber si un conjunto $A$ es superconjunto del conjunto $B$, podemos utilizar el método `.issuperset()` o el comparador `>=`. Para saber si se trata de un superconjunto propio, tenemos el comparador `>`.

```python
A, B = {0, 3, 7, 2, 5}, {2, 3, 0}
A.issuperset(B)
# A es superconjunto de B, es lo mismo que hacer A >= B
```

```python
A >= B
```

```python
A > B
```

**Ojo:** ¡Pero no todos los conjuntos son siempre subconjunto o superconjunto unos de otros!

```python
# conjuntos distintos
X = {1,2,3}
Y = {'a', 'b', 'c'}
```

```python
X <= Y
```

```python
X >= Y
```

### Operaciones con conjuntos

Dados los conjuntos $A$ y $B$, vamos a ver una serie de operaciones que podemos hacer con ellos.

setA.png
setB.png

**Unión.** La unión de dos conjuntos $A$ y $B$ es un nuevo conjunto $A\cup B$ que contiene todos los elementos de $A$ y/o todos los de $B$.
setUnion.png

En `Python`, la unión de dos conjuntos se consigue con la función `|` o bien, con el método `.union()`

```python
A, B = {1, 3, 5, 7, 9}, {0, 2, 4, 6, 8}
# unir con el simbolo |
A | B
```

```python
# podemos usar el método union
A.union(B)
```

```python
B.union(A)
```

**Intersección.** La intersección entre dos conjuntos $A$ y $B$ es un nuevo conjunto $A\cap B$ que contiene todos los elementos comunes de $A$ y $B$.

setIntersection.png

En `Python`, la intersección de dos conjuntos se consigue con la función `&` o bien, con el método `.intersection()`

```python
A, B = {0, 1, 3, 5, 8, 13}, {0, 2, 4, 6, 8}
# interserccion con &
A & B
```

```python
A.intersection(B)
```

```python
# el orden no altera el resultado
B.intersection(A)
```

**Diferencia.** La diferencia entre dos conjuntos $A$ y $B$ es un nuevo conjunto $A - B$ que contiene todos los elementos de $A$ que no están en $B$

setDifference.png

En `Python`, la diferencia de dos conjuntos se consigue con la función `-` o bien, con el método `.difference()`



```python
A, B = {0, 1, 4, 5, 8, 13}, {0, 3, 4, 7, 8}
# quita los elementos repetidos entre ambos, dejando solo lo de A
A - B
```

```python
# El orden si altera los resultados
A.difference(B)
```

```python
B.difference(A)
```

**Diferencia simétrica.** La diferencia entre dos conjuntos $A$ y $B$ es un nuevo conjunto $A \vartriangle B$ que contiene todos los elementos de $A\cup B$ no están en $A\cap B$

$$A\vartriangle B = (A\cup B) - (A\cap B) = (A - B)\cup (B - A)$$

setSimetricDifference.png

En `Python`, la diferencia simétrica de dos conjuntos se consigue con el método `.symmetric_difference()`



```python
A, B = {0, 1, 4, 5, 8, 13}, {0, 3, 4, 7, 8}
# quitamos la intersección
A.symmetric_difference(B)
```

```python
# usando los simbolos
(A-B) | (B-A)
```

```python
# otra forma de representarlo, es la unión sin la intersección
(A | B) - (A & B)
```

### Elementos de un conjunto

Podemos añadir un  elemento a un conjunto con el método `.add()`

```python
# si bien los conjuntos son inmutables, solo se pueden modificar usando sus métodos
set1 = {1, 2, 3, 4, 5}
set1.add(-1)
print(set1)

set1.add(0)
print(set1)
```

Para añadir elementos de otro conjunto en el conjunto actual, podemos usar el método `.update()`

```python
set1 = {1, 2, 3, 4, 5}
set2 = {-1, -2, -3, -4, -5}

print(set1)

# unimos dos cojuntos
set1.update(set2)
print(set1)
```

**Observación.** El método `.update()` nos permite añadir los elementos de un iterable al conjunto actual

```python
set1 = {1, 2, 3, 4, 5}
l = [-1, -2, -3, -4, -5]

# podemos añadir los elementos de una lista
set1.update(l)
print(set1)
```

Podemos averiguar si un elemento pertenece a un conjunto con el operador `in`

```python
set1 = {1, -1, 2, -2, 3, -3}
# con in determinamos si un valor pertenece al conjunto
print(3 in set1)
print(0 in set1)
```

También podemos eliminar elementos de un conjunto con los métodos `.remove()` o `.discard()`

```python
set1 = {"a",  "e", "i", "o", "u"}
# ambos métodos son análogos
set1.remove("a") # retorna un error si el elemento no esta en el conjunto
set1.discard("o") # no retorna error en caso no encuentre el elemento

print("a" in set1)
print("o" in set1)
print(set1)
```

**Observación.** Si intentamos eliminar un elemento que no existe ya de por sí en el conjunto con el método `.remove()`, `Python` nos devolverá error, mientras que con el método `.discard()` no devolverá ningún error.

### Tamaño de un conjunto

Para saber cuántos elementos contiene un conjunto, podemos usar la función `len()`del siguiente modo:

```python
set1 = {1, "a", "i", "o", 2, "e", 4, 3, 5, "u"}
# de igual forma, usamos len para obtener el tamaño del conjunto
print(len(set1))
```

## Bucles y conjuntos

Podemos acceder a todos los elementos de un conjunto mediante un bucle `for`

```python
set1 = {"manzana", "pera", "melón"}
# podemos iterar un conjunto, pero no tenemos un orden especifico
for item in set1:
  print(item)
```

## Más métodos de conjuntos

El método `.pop()` nos devuelve un objeto del conjunto (como no hay orden, no sabemos cuál es) y lo elimina de éste.

```python
set1 = {"u", 5, -3, "a", "b"}
set1
```

```python
# retorna un elemento del conjunto y lo elimina, pero este elemento es aleatorio, porque no hay orden en los elementos
set1.pop()
```

```python
set1
```

El método `.clear()` vacía el conjunto.

```python
# eliminamos todos los elementos del conjunto
set1 = {"a", "e", "i", "o", "u"}
set1.clear()
print(set1)
```

# Ejercicio

Vamos a pedirle al usuario una frase y vamos a guardar en un conjunto las letras que aparecen en dicha frase.

```python
s = input("Introduce una frase: ")

letters = set()

for c in s.lower():
  if c != " ":
    letters.add(c)

print(letters)
```

# Ejercicio

Vamos a pedirle al usuario dos palabras y vamos a calcular la intersección de las letras de cada palabra. Para ello habrá que crear dos conjuntos que contendrán, respectivamente, las letras que forman cada palabra.

```python
word1 = input("Palabra 1: ")
word2 = input("Palabra 2: ")

letters1 = set()
letters2 = set()

# conjunto con cada palabra
for l in word1.lower():
  letters1.add(l)

for l in word2.lower():
  letters2.add(l)

# intersección entre conjuntos
common = letters1 & letters2
print(common)
```

# Ejercicio

Vamos a pedirle 4 números enteros al usuario. Se corresponderán con los extremos de los intervalos $[a, b]$ y $[c, d]$. Vamos a generar dos conjuntos que guarden, respectivamente, los enteros contenidos en cada uno de los intervalos (incluyendo los extremos) y, finalmente, calcularemos la diferencia simétrica.

```python
a = int(input("a = "))
b = int(input("b = "))
c = int(input("c = "))
d = int(input("d = "))

if b < a or d < c:
  print("No has proporcionado dos intevalos")

else:
  # conjuntos con los intervalos
  ab = set(range(a, b + 1))
  cd = set(range(c, d + 1))

  # diferencia simetrica
  print(ab.symmetric_difference(cd))
```

# Ejercicio

Vamos a guardar en un conjunto los números primos comprendidos entre 2 y el número $n$ que nos indique el usuario mediante la criba de [Eratóstenes](https://es.wikipedia.org/wiki/Criba_de_Erat%C3%B3stenes).

440px-Sieve_of_Eratosthenes_animation.gif

```python
n = int(input("Introduce un número entero mayor que 2: "))
primes = set(range(2, n + 1))
numbers = list(range(2, n + 1))
# lista con todas las entradas en true
multiples = [True for x in range(len(numbers))]

for i in range(len(numbers)):
  if multiples[i] == False: # si es false, indica que el número esta tachado
    continue
  for j in range(i + 1, len(numbers)): # calculamos los multiplos con los valores de i en adelante
    if numbers[j] % numbers[i] == 0: # numbers[j] es mayor que numbers[i]
      multiples[j] = False
      # lo eliminamos de la lista de primos
      primes.discard(numbers[j])

# lista final de primos
print(primes)
```

# Ejercicio

Vamos a crear un programa que nos devuelva el elemento máximo de un conjunto sin utilizar la función `max()`.

```python
myset = {20, 3, 5, 10, 9, 30, -5, 30}
# max = max(myset) # de esta forma podemos obtener el mayor del conjunto
max = -99999

for e in myset:
  if e > max:
    max = e

print("El máximo de {} es {}".format(myset, max))
```
